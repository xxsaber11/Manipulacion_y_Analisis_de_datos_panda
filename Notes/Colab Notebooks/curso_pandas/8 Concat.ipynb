```
import pandas as pd
import numpy as np
```

```
pd.options.display.float_format = '{:.2f}'.format
```

```
np.set_printoptions(precision=2)
```

```
# generando una matriz e 2 x 5 con numeros aleatorios
x1 = np.random.rand(2,5) * 10
x1
```

```
x2 = np.random.rand(2,5) * -1
x2
```

**Concatenar ambas matrices con numpy**

```
x1[0]
```

```
np.concatenate([x1,x2]).shape
```

```
# Concatenar en eje columnas
np.concatenate([x1,x2], axis=1).shape
```

**Concatenar ambas matrices con pandas como Series**

```
s1 = pd.Series(x1[0], index = ['a', 'b', 'c', 'd', 'e'])
s1
```

```
s2 = pd.Series(x2[0], index = ['c', 'd', 'e', 'f', 'g',])
s2
```

```
pd.concat([s1,s2])
```

```
pd.concat([s1,s2], axis=1)
```

```
s1.reset_index(drop=True)
```

```
pd.concat([s1.reset_index(drop=True),s2.reset_index(drop=True)], axis=1)
```

**Concatenar Dataframes**

```
df1 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'])
df1
```

```
df2 = pd.DataFrame(np.random.rand(3,2)*10,columns=['a','b'], index=[2,3,4])
df2
```

```
pd.concat([df1,df2])
```

```
# Concatenar  a lo largo de las columnas
pd.concat([df1,df2], axis=1)
```

```
# Utilizando concat con joins
pd.concat([df1,df2], axis=1, join='inner')
```

```
# Borrando los indices
pd.concat([df1.reset_index(drop=True),
          df2.reset_index(drop=True)
          ], axis=1)
```

**Usando append**

```
df1.append(df2)
```

```
df1.append(df2).append(df2)
```

```
# Unir a traves del eje de las columnas usando la Transpuesta
df1.T.append(df2.T)
```
