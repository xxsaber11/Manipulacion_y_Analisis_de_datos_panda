```
#Creando una Pandas Series
sr = pd.Series([10,9,8,7,6])
sr
```

```
#Obteniendo sus valores
sr.values
```

```
#Obteniendo el rango (inicio, fin, pasos)
sr.index
```

```
#la forma de la series
sr.shape
```

```
# Acceder a un elemento
sr[3]
```

```
# Acceder a informacion por medio de los indices
sr[[0,4,2]]
```

```
# Definiendo un nuevo indice para la columna index
sr = pd.Series([10,9,8,7,6], index = ['a','b','c','d','e'])
sr
```

```
# acceder a valores con los nuevos indices
sr['c']
```

```
# acceder a valores con los nuevos indices
sr[['c','d','e']]
```

```
# Usando notacion de slices para valores en un rango
sr['b':'e']
```
