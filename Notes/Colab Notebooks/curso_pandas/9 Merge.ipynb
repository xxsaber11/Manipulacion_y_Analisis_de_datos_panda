```
import pandas as pd
pd.__version__
```

```
df_left = pd.DataFrame(
      {'X':['x0','x1','x2','x3'], 
       'W':['w0','w1','w2','w3'], 
       'Y':['y0','y1','y2','y3'], 
       'Mix':['y2','y3','a2','a3']}, 
       index =[0,1,2,3])
df_left
```

```
df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
df_right
```

```
pd.merge(df_left,df_right)
```

```
# Especificando el tipo de union
pd.merge(df_left,df_right, how='inner', on='Y')
```

```
pd.merge(df_left,df_right, how='inner', left_on='Mix', right_on='Y')
```

```
pd.merge(df_left,df_right, how='inner', left_on='Mix', right_on='A')
```

Ejemplos

```
df_left = pd.DataFrame(
      {'X':['x0','x1','x2','x3'], 
       'W':['w0','w1','w2','w3'], 
       'Y':['y0','y1','y2','y3']}, 
       index =[0,1,2,3])
df_left
```

```
df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
df_right
```

```
pd.merge(df_left, df_right, how='inner', on='Y')
```

```
 pd.merge(df_left, df_right, how='left', on='Y')
```

```
 pd.merge(df_left, df_right, how='right', on='Y')
```

```
 pd.merge(df_left, df_right, how='outer', on='Y')
```

```
# Agrego segunda columna A para explorar otros merge
df_left = pd.DataFrame(
      {'X':['x0','x1','x2','x3'], 
       'W':['w0','w1','w2','w3'], 
       'Y':['y0','y1','y2','y3'],
       'A':['a0','a1','a2','a3']
       }, 
       index =[0,1,2,3])
df_left
```

```
 pd.merge(df_left, df_right, how='outer', on=['Y', 'A'])
```

```
 # Cambiando los subfijos
 pd.merge(df_left, df_right, how='outer', on='A', suffixes=['_left', '_right'])
```
