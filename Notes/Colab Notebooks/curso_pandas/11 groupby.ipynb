```
# Agrupando por sexo
df.groupby('sex').mean()
```

```
# Porcenaje de propina
df['percent_tip'] = df['tip']/df['total_bill']
df
```

```
df.groupby('sex').mean()
```

```
# Usar promedios no siempre es bueno por los outliers de la misma distribucion, usa la mediana mejor
df.groupby('sex').median()
```

```
df.groupby('sex')[['total_bill','percent_tip']].describe()
```

```
def mean_eur2usd(x):
  return np.mean(x) *1.12
```

```
 df.groupby(['sex', 'time'])[['total_bill','percent_tip']].apply(mean_eur2usd)
```

```
 df.groupby(['sex', 'time'])[['total_bill','percent_tip']].apply(np.std)
```

```
# Podemos tomar las funciones predefinidas o utilizar lambdas
df.groupby(['sex', 'time'])[['total_bill','percent_tip']].apply(lambda x: np.mean(x) *1.12)
```

```
# Si necesitas aplicar mas de una funcion, utiliza aggregate, ya sea con una lista o con un diccionario
df.groupby(['sex', 'time'])[['total_bill','percent_tip']].aggregate([np.min, np.mean, np.max])
```

```
# tenemos el shortcut agg
df.groupby(['sex', 'time'])[['total_bill','percent_tip']].agg([np.min, np.mean, np.max])
```

```
# Usando diccionarios
dict_agg = {'tip': [min, max], 'total_bill':[np.mean, mean_eur2usd]}
```

```
df.groupby(['sex','time'])[['total_bill', 'tip']].agg(dict_agg)
```

```
# Definiendo filtros
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

```
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
```

```
df_filtered.groupby(['sex', 'time']).count()
```

## Group By: extraer valor con variables categÃ³ricas

Usando la columna auxiliar ones, obtendremos valor al hacer operaciones sobre esta al agrupar variables categoricas, sin hacer la conversion de variales

```
# Creo una coluna ones
df['ones'] = 1
df
```

```
# Agrupamos por genero y si son fumadores
df_g = df.groupby(['sex', 'smoker'])[['ones']].sum()
df_g
```

```
df_g.groupby(level=0).apply(
    lambda x:
    x /x.sum() * 100
)
```

### Transformando variables numericas a categoricas

```
pd.cut(df['total_bill'], bins=3).value_counts()
```

```
# Definiendo el ancho de los bins (categorias)
pd.cut(df['total_bill'], bins=[2,18,35, 60]).value_counts()
```

```
# Agregando columna bins_total
df['bins_total'] = pd.cut(df['total_bill'], bins=[2,18,35, 60])
df
```

```
# obtenemos los porcentajes
df.groupby(['time', 'bins_total'])[['ones']].count().groupby(level=0).apply(
    lambda x:
    x /x.sum() * 100
)
```

## Pivot Table

```
# Precio medio de las comidas
df.groupby(['sex','time'])[['total_bill']].mean()
```

```
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

```
df_gp.pivot_table(values='total_bill', index='sex', columns='time')
```

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)


```
# Aplicamos pivot_table al DataFrame Original, por defecto calcula tambien la media de los valores
df.pivot_table(values='total_bill', index='sex', columns='time')
```

```
# `aggfunc` permite cambiar el tipo de operacion al que nosotros querramos
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=np.median)
```

```
# Puedo agregar tambien un array o un diccionario
df_pivot  = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median, np.std])
df_pivot
```

```
# Deshacer las columnas y rehacer un DataFrame con `unstack()`
df_pivot.unstack().reset_index().to_frame
```

```
df.pivot_table(values=['total_bill', 'tip'], index='day', columns='time', aggfunc=[np.mean, np.median, np.std])
```
