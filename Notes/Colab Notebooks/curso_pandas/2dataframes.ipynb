```
import numpy as np
```

```
dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
dict_data
```

```
# Creando un DataFrame con un index por nombres
df = pd.DataFrame(dict_data, index = ['ana','benito','camilo','daniel','erika','fabian','gabriela'])
df
```

```
df.index
```

```
df.columns
```

```
df.values
```

```
# Accediendo a los valores de la columna (series)
df['edad']
```

```
# Accediendo a varias columnas
df[['edad', 'cm', 'Q1']]
```

```
# Accediendo solo a los datos de Ana para las columnas anteriores
df.loc['ana',['edad', 'cm', 'Q1']]
```

```
# iloc
df.iloc[2,1]
```

```
df.iloc[[2,4,5], [1,3]]
```

```
df.iloc[:,[1,3]]
```

```
# Realizando operaciones
# filtrando 
df[df['edad'] >= 12]
```

```
# And 
df[(df['edad'] >= 12) & (df['pais'] =='mx')]
```

```
df.query('edad>12')
```

```
df[df['Q2'] >= df['Q1']]
```
